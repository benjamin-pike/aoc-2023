```python
with open('../inputs/21.txt') as f:
    data = f.read().splitlines()
```

```python
def dfs(location, garden_map, step_n, end_step, endpoints, memo):
    if (location, step_n) in memo:
        return
    
    memo.add((location, step_n))
    
    if step_n == end_step:
        endpoints.add(location)
        return
        
    y1, x1 = location
    for y2, x2 in [(y1 + 1, x1), (y1 - 1, x1), (y1, x1 + 1), (y1, x1 - 1)]:
        if (
            not 0 <= y2 < len(garden_map) or 
            not 0 <= x2 < len(garden_map[0]) or
            garden_map[y2][x2] == '#' 
        ):
            continue
            
        dfs((y2, x2), garden_map, step_n + 1, end_step, endpoints, memo)  

def count_endpoints_dfs(garden_map, steps):
    start = next((y, x) for y, row in enumerate(data) for x in range(len(row)) if data[y][x] == 'S')

    endpoints = set()
    memo = set()
    dfs(start, garden_map, 0, steps, endpoints, memo)
    
    return len(endpoints)
```

```python
# Part 1
count_endpoints_dfs(data, 64)
```

```python
# Part 2  
def bfs(start, garden_map, total_steps):
    queue = [(*start, total_steps)]
    seen = set()
    endpoints = set()
    
    while queue:
        y1, x1, remaining_steps = queue.pop(0)
        
        if remaining_steps % 2 == 0:
            endpoints.add((y1, x1))
        if remaining_steps == 0:
            continue
                    
        for y2, x2 in [(y1 + 1, x1), (y1 - 1, x1), (y1, x1 + 1), (y1, x1 - 1)]:
            if (
                not 0 <= y2 < len(garden_map) or 
                not 0 <= x2 < len(garden_map[0]) or
                garden_map[y2][x2] == '#' or
                (y2, x2) in seen
            ):
                continue
            
            queue.append((y2, x2, remaining_steps - 1))
            seen.add((y2, x2))
            
    return len(endpoints) 

def count_endpoints_bfs(garden_map, steps):
    start = next((y, x) for y, row in enumerate(data) for x in range(len(row)) if data[y][x] == 'S')
    endpoints = bfs(start, garden_map, steps)
    
    return endpoints 

def calc_large_steps(garden_plot, steps):
    nr, nc = [len(garden_plot), len(garden_plot[0])]

    assert (nr == nc) # map is square

    width = nr
    middle = width // 2

    assert ('#' not in data[middle]) # start row is clear
    assert ('#' not in [row[middle] for row in garden_plot]) # start column is clear

    odd_p = count_endpoints_bfs(garden_plot, width) # odd plots
    even_p = count_endpoints_bfs(garden_plot, width - 1) # even plots

    tw = nr # tile width
    tr = tw // 2 # tile radius
    utr = steps // tw # universe tile radius

    assert (steps == tr + (utr * tw))

    odd_t = (utr + 1) ** 2 # odd_tiles
    even_t = (utr) ** 2 # even_tiles

    odd_c = odd_p - count_endpoints_bfs(garden_plot, tr) # odd corners
    even_c = even_p - count_endpoints_bfs(garden_plot, tr - 1) # even corners

    return (odd_t * odd_p) + (even_t * even_p) - ((utr + 1) * odd_c) + (utr * even_c)

calc_large_steps(data, 26501365)
```
